```python
import pandas as pd
```

```python
file = open('./Data/result2')
table = pd.DataFrame(columns=["id","name","location"])
```

```python
i = 0 
while i<=1999:
    id_ = file.readline().replace('ID: ', '').replace('\n', '')
    name = file.readline().replace('Name: ', '').replace('\n', '')
    location = file.readline().replace('Location: ', '').replace('\n', '')
    # Pass the row elements as key value pairs to append() function 
    table = table.append({'id' : id_ , 'name' : name, 'location': location} , ignore_index=True)
    i += 1
```
